```python
import pandas as pd
import requests
```

```python
url="https://v6.exchangerate-api.com/v6/0e0ce298ff97b0e9b38cb3ef/latest/USD"
```

```python
response=requests.get(url)
```

```python
if response.status_code==200:
    data=response.json() #we get in dictionary format
```

```python
data
```

```python
rates=data["conversion_rates"]
rates.items()# it gives a list of tuples with kiey and value pairs
```

```python
conversion_table=pd.DataFrame(rates.items(),columns=["currecy","current exchange rate"])
```

```python
conversion_table
```

```python
conversion_table.to_csv("conversion.csv")
```
